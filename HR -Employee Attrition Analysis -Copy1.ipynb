### <font color=purple>Is Attrition age  dependent?</font> <br>


```python
age_att=emp_df.groupby(['Age','Attrition'])['Attrition'].count().reset_index(name='Counts')
fig = px.line(age_att[age_att['Attrition']=='Yes'],x='Age',y='Counts',color='Attrition',title='Agewise Counts of Attrition Employees in an Organization')
pio.show(fig)
```

<font color=purple>__Observation__</font> <br>
Attrition Employees fall under the age between 28 and 32<br>
Attrition count decreases as age increases<br>
The Probability of employees leaving the company between 18 and 20 is much less<br>

### <font color = purple> __Is Income an accountable reason for Attrition?__</font><br>

```python
inc_att = emp_df.groupby(['MonthlyIncome','Attrition'])['Attrition'].count().reset_index(name = 'Counts')
inc_att['MonthlyIncome']=round(inc_att['MonthlyIncome'],-3)
inc_att =inc_att.groupby(['MonthlyIncome','Attrition'])['Attrition'].count().reset_index(name = 'Counts')
fig =px.line(inc_att,x='MonthlyIncome',y='Counts',color='Attrition',title='Income based Attrition Employees in an Organization')

pio.show(fig)
```

```python
emp_df['MonthlyIncome']
```

```python
per_att = emp_df.groupby(['PercentSalaryHike','Attrition'])['Attrition'].count().reset_index(name='Counts')

fig = px.bar(per_att,x='PercentSalaryHike',y ='Counts',color ='Attrition',text_auto =True,title='Employee Attrition based on percentsalaryhike')
pio.show(fig)
```

<font color = purple>__Observation__</font> <br>
Among 237 attrition employee count,more than 50% of the employee left the company with much less salary hike between 11 to 14%.<br> Percent Salaryhike plays an important role in employees turnover.

### <font color = purple>__Do Satisfactory factors such as Environment, Job and Colleague Relationships reflect in EmployeeAttritionRate?__</font><br> 

```python
envi_att = emp_df.groupby(['EnvironmentSatisfaction','Attrition'])['Attrition'].count().reset_index(name='Counts')


fig= px.area(envi_att,x ='EnvironmentSatisfaction',y ='Counts',color ='Attrition',text='Counts',title = 'Environment Satisfaction Rating Count of Employees')

pio.show(fig)
```

<font color =purple>__Observation__</font><br>
As expected,Low Environment satisfacting rating(1 -2) has high Attrition Employees<br>
Organization need to focus more on employees who falls under the rating  2 to 3 as<br>
there are great chances of employees to leave the company

```python
Job_att = emp_df.groupby(['JobSatisfaction','Attrition'])['Attrition'].count().reset_index(name='Counts')


fig= px.area(Job_att,x ='JobSatisfaction',y ='Counts',color ='Attrition',text='Counts',title = 'Job Satisfaction Rating Count of Employees')

pio.show(fig)
```

### <font color =purple>Genderwise Attrition Rate</font><br>
